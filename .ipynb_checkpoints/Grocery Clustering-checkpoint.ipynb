Create a pivot table to show how groceries are purchased at each instance. 

```python
full['bought'] = 1
```

```python
purchase_table = full.pivot_table(index = ['user_id'], columns = ['item_name'], values = 'bought')
purchase_table.head()
```

```python
# Fill NA with 0
purchase_table.fillna(0, inplace = True)
```

The above line of code will transform the data into binary (0 or 1). 1 will mean that the user bought the item and 0 will mean that the user did not buy the item. This is done because typically, all data processed by a computer is converted to binary.

```python
purchase_table.head()
```

```python
# Transpose the table to make item_name the main column since our goal is to evaluate how items are bought together. 
ptt = purchase_table.transpose()
ptt
```

```python
# Perform feature scaling
ptt_scaled = StandardScaler().fit_transform(ptt)
```

```python
ptt_scaled
```

### Perform PCA and Clustering 

```python
pca = PCA()
```

```python
ptt_pca = pca.fit(ptt_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(ptt_scaled)
```

```python
ptt_pca_scores = pca.transform(ptt_scaled)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
ptt_rescaled = scaler.fit_transform(ptt)
```

```python

pca = PCA(n_components = 0.95)
pca.fit(ptt_rescaled)
reduced = pca.transform(ptt_rescaled)
```

```python
#Using elbow methid to determine the number of clusters
wcss = []
for i in range(1, 11):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(reduced)
    wcss.append(kmeans_pca.inertia_)
plt.plot(range(1, 11), wcss, marker = 'x')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Create a new dataframe with the assigned clusters based on the ML algorithm
new_df = pd.concat([ptt.reset_index(drop = True), pd.DataFrame(ptt_pca_scores)], axis = 1)
new_df.columns.values[-2: ] = [1,2]
new_df['Assigned Cluster'] = kmeans_pca.labels_
```

```python
new_df = new_df.reset_index().set_index(ptt.index)
```

```python
# Create a new column called Cluster Tag and map the clusters to string representatives
new_df["Cluster Tag"] = new_df['Assigned Cluster'].map({0:'first', 1:'second', 2:'third', 3:'fourth', 4: 'fifth'})
```

```python
# Preview new dataframe
new_df.head()
```
